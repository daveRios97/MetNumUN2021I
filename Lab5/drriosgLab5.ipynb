## **1. Netflix**

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2019, 1, 1)
        self.SetEndDate(2121, 1, 1)

        #1,2. Select IWM minute resolution data and set it to Raw normalization mode
        self.nflx = self.AddEquity("NFLX", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)

    def OnData(self, data):

        #3. Place an order for 100 shares of IWM and print the average fill price
        if not self.Portfolio.Invested:
            self.MarketOrder("NFLX", 100)
            self.Debug( str(self.Portfolio["NFLX"].AveragePrice))
```

## **Portafolio Completo**

Ahora creamos un portafolio completo con las empresas anteriormente mencionadas. Inicializando el SetCash con un Millon de dolares.

```python
def Initialize(self):
        self.SetStartDate(2019, 4, 1)  # Set Start Date
        self.SetEndDate(2121, 4, 1) #Set End Date
        self.SetCash(1000000)  # Set Strategy Cash
        
        
        self.nflx = self.AddEquity("NFLX", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.tsla = self.AddEquity("TSLA", Resolution.Minute)
        self.tsla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.amzn = self.AddEquity("AMZN", Resolution.Minute)
        self.amzn.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.goog = self.AddEquity("AAPL", Resolution.Minute)
        self.goog.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.aapl = self.AddEquity("GOOG", Resolution.Minute)
        self.aapl.SetDataNormalizationMode(DataNormalizationMode.Raw)
```
